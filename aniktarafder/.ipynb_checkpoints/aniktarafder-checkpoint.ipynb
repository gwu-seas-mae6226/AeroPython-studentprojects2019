# Determination of pressure distribution around three-cornered arbitrary bodies

Low-aspect-ratio wings and large volume
fuselages with long nose lengths increases the importance of the
fuselage contribution to the aerodynamic characteristics of aircraft. In addition, engine installations and wing-fuselage interference
have led to a variety of fuselage cross sections. Cross section can have considerable effect on both
the longitudinal and lateral aerodynamic characteristics, especially at
high angles of attack. Due to increased fuselage length ahead of both the center of
gravity and the wing, it might be expected that the spin characteristics
of the aircraft would be affected to an even greater extent by changes
in forebody cross section. Thus, a better understanding of the effect of cross section on the fuselage
forces, especially at combined angles of attack and sideslip, is important to the aircraft designer.Besides, landing in a strong crosswind can be a challenging and in some cases downright dangerous. There are many techniques used during the approach and landing phase which center around the crab and sideslip approach.  The crab and sideslip are the primary methods used when crosswinds are stronger than usual.

The flow past the body is resolved into
two components, one parallel to the body axis and one normal to it. It
is then assumed that the crossflow is independent of the axial flow and
therefore related to the flow about a cylinder at right angles to the

in spins, the problems involving the degree of interdependence between
the crossflow and axial flow are largely eliminated. For
both these methods, it is necessary to have a knowledge of the forces on

## Conformal mapping to transform a circle into three cornered bodies using singular points

Two-dimensional ideal fluid flow about a great variety of cross-sectional shapes can be obtained by a conformal transformation of the flow about a circle. Finding the equation $\xi=f(z)$ that will transform a circle into a given shape can be fascilitated by consideration of singular points. The singular points of a transformation $\xi=f(z)$ are those values of $z$ for which $\frac{d\xi}{dz} = 0$. Thus, for
example, the transformation

$$\xi = z-\frac{1}{2 z^2}$$
has singular points $z=-1$, $e^{i\pi/3}$, $e^{-i\pi/3}$. Since, 

$$\frac{d\xi}{dz} = 1 + \frac{1}{z^3}=\left( 1+\frac{1}{z} \right) \left( 1-\frac{e^{i\pi /3}}{z} \right) \left( 1-\frac{e^{-i\pi /3}}{z} \right) 
$$

Application of this transformation to a circle $|z|=1$ passing through the three singular points
produces a triangular-like shape with concave
sides and three vertices corresponding to the three singular points as shown in figure bellow,
![Figure](figures/p23.png)
<center>Ref[1]</center>

### Basic steps:

(1) A transformation function of the form

$$
\xi =z+\frac{b_1}{z}+\frac{b_2}{z^2}+...\,\,...\,\,...+\frac{b_n}{z^n}
$$
is considered.

(2) Points for which $\frac{d\xi}{dz} = 0$ (singular points) occures are derived which determines $b_1$, $b_2$,... ... ... $b_n$ constants.

Many fuselage cross sections are characterized by three "corners" (that is, three places where the curvature is a maximum). As a means of approximating such shapes, a three-singular-point transformation of a circle should first be tried. More exact, approximations can be made by using more singular points; however, each added point produces another term in the transformation function. Here, for three corner points the transformation function that would be used is of the form,

$$
\xi =z+\frac{b_1}{z}+\frac{b_2}{z^2}
$$

which will finally lead to an equation of the form,

$$
\xi =z-\frac{1-l^2}{z}-\frac{l}{2z^2}
$$
where $b_1$ and $b_2$ are constants and for our present case $b_1=-(1-l^2)$ and $b_2=-l/2$.
The equation of the circle is, 
$$
(x-a)^2+ y^2 = R^2
$$
Thus $R$, $a$ and $l$ determines the final shape of the body. [1] gives elaborate description and derivation of this method.
![Figure](figures/24.png)
<center>Ref[1]</center>

The velocitites in two reference frames are related by,

$$V_\xi = V_z/\frac{d\xi}{dz}$$ where, $$\frac{d\xi}{dz} =1+\frac{1-l^2}{z^2}+\frac{l}{z^3}$$

A source panel method would be used to determine the pressure distribution around the surface.

```python
import numpy
from matplotlib import pyplot
import math
from scipy import integrate
%matplotlib inline
```

```python
################################################## Functions and Classes ########################################################

#Transformation of circle into three cornered bodies
def circle(xc,yc,R,N):
    theta = numpy.linspace(0,2*math.pi, N)
    x, y= R*numpy.cos(theta)+xc, R*numpy.sin(theta)+yc
    return x,y

def polhamus_transform(z,beta):
    l = 2*numpy.cos(beta)
    xi = z-(1-l**2)/z-l/(2*z**2)
    
    return xi

def stream_doublet(kappa, xd, yd, X, Y):
      
        psi = - kappa / (2 * math.pi) * (Y - yd) / ((X - xd)**2 + (Y - yd)**2)
        return psi
def vel_doublet(kappa, xd, yd, X, Y):
     
    u = (- kappa / (2 * math.pi) * ((X - xd)**2 - (Y - yd)**2) / ((X - xd)**2 + (Y - yd)**2)**2)
    v = (- kappa / (2 * math.pi) * 2 * (X - xd) * (Y - yd) / ((X - xd)**2 + (Y - yd)**2)**2)
    return u, v

def polhamus_derivative(z, beta):
     
    l = 2*numpy.cos(beta)
    dxidz = 1 + (1-l**2)/z**2 + l/z**3
    
    return dxidz

class Panel:
  
    def __init__(self, xa, ya, xb, yb):
  
        self.xa, self.ya = xa, ya
        self.xb, self.yb = xb, yb
        
        self.xc, self.yc = (xa + xb) / 2, (ya + yb) / 2  # control-point (center-point)
        self.length = math.sqrt((xb - xa)**2 + (yb - ya)**2)  # length of the panel
        
        # orientation of the panel (angle between x-axis and panel's normal)
        if xb - xa <= 0.0:
            self.beta = math.acos((yb - ya) / self.length)
        elif xb - xa > 0.0:
            self.beta = math.pi + math.acos(-(yb - ya) / self.length)
        
        # location of the panel
        if self.beta <= math.pi:
            self.loc = 'upper'
        else:
            self.loc = 'lower'
        
        self.sigma = 0.0  
        self.vt = 0.0  
        self.cp = 0.0 
        
def define_panels(x, y, N=40):
  
    R = (x.max() - x.min()) / 2  
    x_center = (x.max() + x.min()) / 2  
    
    x_circle = x_center + R * numpy.cos(numpy.linspace(0.0, 2 * math.pi, N + 1))
    
    x_ends = numpy.copy(x_circle) 
    y_ends = numpy.empty_like(x_ends)  

    x, y = numpy.append(x, x[0]), numpy.append(y, y[0])  
    
    
    I = 0
    for i in range(N):
        while I < len(x) - 1:
            if (x[I] <= x_ends[i] <= x[I + 1]) or (x[I + 1] <= x_ends[i] <= x[I]):
                break
            else:
                I += 1
        a = (y[I + 1] - y[I]) / (x[I + 1] - x[I])
        b = y[I + 1] - a * x[I + 1]
        y_ends[i] = a * x_ends[i] + b
    y_ends[N] = y_ends[0]
    
    panels = numpy.empty(N, dtype=object)
    for i in range(N):
        panels[i] = Panel(x_ends[i], y_ends[i], x_ends[i + 1], y_ends[i + 1])
    
    return panels

class Freestream:
   
    def __init__(self, u_inf=1.0, alpha=0.0):
     
        self.u_inf = u_inf
        self.alpha = numpy.radians(alpha)  
        
def integral(x, y, panel, dxdz, dydz):
   
    def integrand(s):
        return (((x - (panel.xa - math.sin(panel.beta) * s)) * dxdz +
                 (y - (panel.ya + math.cos(panel.beta) * s)) * dydz) /
                ((x - (panel.xa - math.sin(panel.beta) * s))**2 +
                 (y - (panel.ya + math.cos(panel.beta) * s))**2) )
    return integrate.quad(integrand, 0.0, panel.length)[0]

def build_matrix(panels):
   
    N = len(panels)
    A = numpy.empty((N, N), dtype=float)
    numpy.fill_diagonal(A, 0.5)
    
    for i, p_i in enumerate(panels):
        for j, p_j in enumerate(panels):
            if i != j:
                A[i, j] = 0.5 / math.pi * integral(p_i.xc, p_i.yc, p_j,
                                                   math.cos(p_i.beta),
                                                   math.sin(p_i.beta))
    
    return A

def build_rhs(panels, freestream):
   
    b = numpy.empty(len(panels), dtype=float)
    
    for i, panel in enumerate(panels):
        b[i] = -freestream.u_inf * math.cos(freestream.alpha - panel.beta)
    
    return b

def get_tangential_velocity(panels, freestream):
   
    N = len(panels)
    A = numpy.empty((N, N), dtype=float)
    numpy.fill_diagonal(A, 0.0)
    
    for i, p_i in enumerate(panels):
        for j, p_j in enumerate(panels):
            if i != j:
                A[i, j] = 0.5 / math.pi * integral(p_i.xc, p_i.yc, p_j,
                                                   -math.sin(p_i.beta),
                                                   math.cos(p_i.beta))
    
    b = freestream.u_inf * numpy.sin([freestream.alpha - panel.beta 
                                      for panel in panels])
    
    sigma = numpy.array([panel.sigma for panel in panels])
    
    vt = numpy.dot(A, sigma) + b
    
    for i, panel in enumerate(panels):
        panel.vt = vt[i]
        
def get_pressure_coefficient(panels, freestream):
   
    for panel in panels:
        panel.cp = 1.0 - (panel.vt / freestream.u_inf)**2
```

```python
##################################################### Geometry #################################################################

N = 101
R=1.0
R1=1.0
xc, yc = 0.0, 0.0
R1=1.0
x, y = circle(xc,yc, R1, N)
z1= x +y*1j
beta1 = 20
kt1 = polhamus_transform(z1,beta1)

width = 16
height = 16
pyplot.figure(figsize=(width, height))
pyplot.subplot(2,2,1)
pyplot.title('$R=1.0\,\\beta=20$')
pyplot.grid(True)
pyplot.xlabel('x', fontsize=16)
pyplot.ylabel('y', fontsize=16)
pyplot.xlim(-R-1, R+1)
pyplot.ylim(-R-1, R+1)
#pyplot.subplot(2, 1, 1)
pyplot.plot(z1.real,z1.imag, color = 'g', linewidth=0.1, marker='.')
pyplot.plot(kt1.real,kt1.imag,color = 'b', linewidth=0.1, marker='.')

#idx_te = numpy.argmax(kt.real, axis=None)
#x_te, y_te = kt.real[97], kt.imag[97]

#x_first, y_first = kt.real[0], kt.imag[0]
R2=1.4
x, y = circle(xc,yc, R2, N)
z2= x +y*1j
beta2 = 20
kt2 = polhamus_transform(z2,beta2)
pyplot.subplot(2,2,2)
pyplot.title('$R=1.4\,\\beta=20$' )
pyplot.grid(True)
pyplot.xlabel('x', fontsize=16)
pyplot.ylabel('y', fontsize=16)
pyplot.xlim(-R-1, R+1)
pyplot.ylim(-R-1, R+1)
#pyplot.subplot(2, 1, 1)
pyplot.plot(z2.real,z2.imag, color = 'g', linewidth=0.1, marker='.')
pyplot.plot(kt2.real,kt2.imag,color = 'b', linewidth=0.1, marker='.')

R3=1.0
x, y = circle(xc,yc, R3, N)
z3= x +y*1j
beta3 = 2
kt3 = polhamus_transform(z3,beta3)
pyplot.subplot(2,2,3)
pyplot.title('$R=1.0\,\\beta=2$')
pyplot.grid(True)
pyplot.xlabel('x', fontsize=16)
pyplot.ylabel('y', fontsize=16)
pyplot.xlim(-R-1, R+1)
pyplot.ylim(-R-1, R+1)
#pyplot.subplot(2, 1, 1)
pyplot.plot(z3.real,z3.imag, color = 'g', linewidth=0.1, marker='.')
pyplot.plot(kt3.real,kt3.imag,color = 'b', linewidth=0.1, marker='.')

R4=1.4
x, y = circle(xc,yc, R4, N)
z4= x +y*1j
beta4 = 2
kt4 = polhamus_transform(z4,beta4)
pyplot.subplot(2,2,4)
pyplot.title('$R=1.4\,\\beta=2$')
pyplot.grid(True)
pyplot.xlabel('x', fontsize=16)
pyplot.ylabel('y', fontsize=16)
pyplot.xlim(-R-1, R+1)
pyplot.ylim(-R-1, R+1)
#pyplot.subplot(2, 1, 1)
pyplot.plot(z4.real,z4.imag, color = 'g', linewidth=0.1, marker='.')
pyplot.plot(kt4.real,kt4.imag,color = 'b', linewidth=0.1, marker='.')
```

```python
Nr, N8 = 101, 101
rad = numpy.linspace(R2,R2+5, Nr)
theta = numpy.linspace(0,2*math.pi, Nr)
RAD, THETA = numpy.meshgrid(rad, theta)
X, Y= RAD*numpy.cos(THETA)+xc, RAD*numpy.sin(THETA)+yc
z = X + Y*1j
kt = polhamus_transform(z, beta2)
Xkt, Ykt = kt.real, kt.imag

width = 16
height = 8
pyplot.figure(figsize=(width, height))
pyplot.subplot(1,2,1)
pyplot.title('z plane')
pyplot.grid(True)
pyplot.xlabel('x', fontsize=16)
pyplot.ylabel('y', fontsize=16)
pyplot.xlim(-6*R2, 6*R2)
pyplot.ylim(-6*R2, 6*R2)
pyplot.scatter(X,Y, color='b', marker = '.')

pyplot.subplot(1,2,2)
pyplot.title('$\u03B5$ plane')
pyplot.grid(True)
pyplot.xlim(-6*R2, 6*R2)
pyplot.ylim(-6*R2, 6*R2)
pyplot.xlabel('$\u03B5_x$', fontsize=16)
pyplot.ylabel('$\u03B5_y$', fontsize=16)
pyplot.scatter(Xkt, Ykt, color='r', marker='.')
```

### Non-zero incident angel

A way to achieve a uniform flow with an inlet angle is by simply rotating the complex plane $z$ to get a new complex plane $z$, where the $x'$-axis (i.e., real part of $z'$) is parallel to the free-stream direction, with its origin located at the center of the cylinder $\left(x_c, y_c\right)$, as shown in the figure below.

The relationship between the $z$-plane and the $z'$-plane is:

$$
\begin{equation}
    z'=\left[ z-(x_c+iy_c) \right]e^{-i\times AoA}
\end{equation}
$$

Now, we can obtain the flow over a cylinder on the new plane $z'$ by adding a free stream with **zero inlet angle** to a doublet **centered at the origin**.
Then, we can obtain the flow on the $z$-plane and, finally, on the $\xi$-plane.
The stream function remains the same at the same point under the three coordinate systems ($z'$, $z$, and $\xi$).

The velocity vector needs to be rotated back from the $z'$-plane to the $z$-plane.

$$
W_z = W_{z'} e^{-i \times \text{AoA}}
$$

```python
############################### Non-zero angle of attack######################
aoa = 20 * math.pi /180


X_prm=  X
Y_prm = Y
z_prm = X_prm + Y_prm*1j
z_new = (z_prm)*numpy.exp(aoa*1j)
X_new = numpy.real(z_new)
Y_new = numpy.imag(z_new)


shape_prm = polhamus_transform(z_new,beta)

width = 16
height = 8
pyplot.figure(figsize=(width, height))
pyplot.subplot(1,2,1)
pyplot.grid(True)
pyplot.xlim(-6, 6)
pyplot.ylim(-6, 6)
pyplot.xlabel('x', fontsize=16)
pyplot.ylabel('y', fontsize=16)
pyplot.contour(numpy.real(shape_prm), numpy.imag(shape_prm), psi,
              levels=numpy.linspace(numpy.min(psi), numpy.max(psi), 51), colors='b', linewidths=1, linestyles='solid')
pyplot.contour(numpy.real(shape_prm), numpy.imag(shape_prm), psi,
              levels=0, colors='r', linewidths=2, linestyles='solid')
#pyplot.scatter(numpy.real(shape_prm), numpy.imag(shape_prm), color='r', marker='.')

#############################################################Vector##############################################################

xd_prm, yd_prm = 0.0,0.0
z_prm = (z)*numpy.exp(-1j*aoa) 
X_prm, Y_prm = numpy.real(z_prm), numpy.imag(z_prm)

u_doublet_prm, v_doublet_prm = vel_doublet(kappa, xd_prm, yd_prm, X_prm, Y_prm)

u_prm = u_freestream + u_doublet_prm
v_prm = v_freestream + v_doublet_prm

wz_prm = u_prm - v_prm*1j
wz_new = wz_prm * numpy.exp(-aoa*1j)
u_new, v_new = numpy.real(wz_new), -numpy.imag(wz_new)

vel_aero = polhamus_derivative(z,beta2)
w_zeta_aero = wz_new/vel_aero

uj_aero, vj_aero = numpy.real(w_zeta_aero), -numpy.imag(w_zeta_aero)

#pyplot.figure(figsize=(width, height))
pyplot.grid(True)
pyplot.subplot(1,2,2)
pyplot.xlim(-3, 3)
pyplot.ylim(-3, 3)
pyplot.xlabel('x', fontsize=16)
pyplot.ylabel('y', fontsize=16)
XJ, YJ = numpy.real(shape_prm), numpy.imag(shape_prm)
pyplot.quiver(XJ[::1, ::1], YJ[::1, ::1], uj_aero[::1, ::1], vj_aero[::1, ::1], scale=100)
```

```python
U_mag = numpy.sqrt(u_new**2+v_new**2)
U_mag_zeta = numpy.sqrt(uj_aero**2+vj_aero**2)
cp = 1- (U_mag/u_inf)**2
cp_zeta = 1- (U_mag_zeta/u_inf)**2

width = 6
height = 6
pyplot.figure(figsize=(width, height))
pyplot.title('$\u03B5$ plane')
pyplot.xlim(-3, 3)
pyplot.ylim(-3, 3)
pyplot.xlabel('$\u03B5_x$', fontsize=16)
pyplot.ylabel('$\u03B5_y$', fontsize=16)
pyplot.contourf(XJ, YJ, cp_zeta,levels=numpy.linspace(-1, 1, 100), extend='both')

cbar=pyplot.colorbar(orientation='horizontal', fraction=0.05)
cbar.set_label('$C_p$', fontsize=16)
cbar.set_ticks([-1.0,-0.5, 0.0,0.5, 1.0])
```

### Source Panel Method

A source-strength distribution would be used to produce potential flow around a given geometry. With a defined body geometry, the strength of sources would be calculated which is needed in each panel to represent a dividing streamline right on the body boundary i.e., the velocity be tangent there. Enforcing the flow-tangency condition on each *control point* approximately makes the body geometry correspond to a dividing streamline (and the approximation improves if we represent the body with more and more panels). So, for each panel $i$, we make $u_n=0$ at $(x_{c_i},y_{c_i})$. Ref [3] gives detailed description and implementation about source panel method.

```python
x,y = kt2.real, kt2.imag
N = 40                        
panels = define_panels(x, y, N)

width = 6
pyplot.figure(figsize=(width, width))
pyplot.grid()
pyplot.xlabel('x', fontsize=16)
pyplot.ylabel('y', fontsize=16)
pyplot.plot(x, y, color='k', linestyle='-', linewidth=2)
pyplot.plot(numpy.append([panel.xa for panel in panels], panels[0].xa),
            numpy.append([panel.ya for panel in panels], panels[0].ya),
            linestyle='-', linewidth=1, marker='o', markersize=6, color='#CD2305')
pyplot.axis('scaled', adjustable='box');
```

#### Discussion

Fuselage cross section and flow incident angle have large contribution on pressure distribution around a non circular cross section. Potential flow solutions provide a ground to understand the flow behavior and pressure distribution around fuselage surface which will help further to understand force distribution around arbitrary bodies. Moreover, conformal mapping is a great tool to obtain desired shapes from a circle which is rather easy to implement.In this present work, non-uniform pressure distribution around the surface indicates a strong interdependence of fuselage body force and cross wind incident angle which in very important during cross-wind landing. Non-uniform pressure distribution also indicates a non-uniform force distribution around fuselage surface during cross-wind landing.
